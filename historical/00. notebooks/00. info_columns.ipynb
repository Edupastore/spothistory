<center><h1>Info columns</h1></center>

- Información de las columnas del conjunto de datos (histórico de reproducciones de Spotify). En la siguiente tabla se explican los campos técnicos:

| Campo técnico                           | Contenido                                                                                                                |
|----------------------------------------|--------------------------------------------------------------------------------------------------------------------------|
| ts                                     | Marca de tiempo UTC de cuándo se dejó de reproducir la canción. Formato: año-mes-día + hora en 24h.                      |
| username                               | Nombre de usuario de Spotify.                                                                                             |
| platform                               | Plataforma usada para escuchar la canción (ej. Android, Google Chromecast).                                              |
| ms_played                              | Duración de la reproducción en milisegundos.                                                                            |
| conn_country                           | Código de país donde se escuchó la canción (ej. SE para Suecia).                                                        |
| ip_addr_decrypted                       | Dirección IP registrada al reproducir la canción.                                                                       |
| user_agent_decrypted                    | Agente de usuario usado para reproducir la canción (ej. navegador como Mozilla Firefox o Safari).                       |
| master_metadata_track_name              | Nombre de la pista.                                                                                                      |
| master_metadata_album_artist_name       | Nombre del artista, grupo o pódcast.                                                                                    |
| master_metadata_album_album_name        | Nombre del álbum que incluye la canción.                                                                                |
| spotify_track_uri                       | URI de Spotify que identifica de forma única la canción (formato: `spotify:track:<base-62 string>`).                     |
| episode_name                            | Nombre del episodio del pódcast.                                                                                        |
| episode_show_name                       | Nombre del programa del pódcast.                                                                                        |
| spotify_episode_uri                      | URI de episodio de Spotify que identifica de forma única el episodio (formato: `spotify:episode:<base-62 string>`).     |
| reason_start                            | Motivo por el que se empezó a reproducir la canción (ej. “trackdone”).                                                |
| reason_end                              | Motivo por el que se acabó de reproducir la canción (ej. “endplay”).                                                  |
| shuffle                                 | Indica si se usó el modo aleatorio (“True” o “False”).                                                                |
| skipped                                 | Indica si los usuarios se saltaron la canción.                                                                         |
| offline                                 | Indica si la canción se escuchó sin conexión (“True” o “False”).                                                      |
| offline_timestamp                        | Marca de tiempo de cuándo se usó el modo sin conexión, si aplica.                                                     |
| incognito_mode                           | Indica si la canción se escuchó durante una sesión privada (“True” o “False”).                                        |
