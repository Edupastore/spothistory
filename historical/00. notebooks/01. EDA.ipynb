```python
for i, col in enumerate(stream.columns):
  print(i, col)
```

```python
# Índices que quiero conservar:
indices = [0, 5, 6, 7, 8]
```

```python
# Filtrar columnas por índice
columnas_seleccionadas = [col for i, col in enumerate(stream.columns) if i in indices]
```

```python
print(columnas_seleccionadas)
```

```python
# Todas las columnas menos columnas_seleccionadas son las que quiero borrar
drop_cols = [col for col in stream.columns if col not in columnas_seleccionadas]
```

```python
# Columnas que vamos a eliminar:
print(drop_cols)
```

```python
# Registros que vamos a eliminar:
drop_rows = stream[stream['spotify_track_uri'].isnull()].index
```

```python
# Eliminamos las columnas y filas para conformar un nuevo dataframe:
st = drop_from_df(stream, index=drop_rows, columns=drop_cols)
```

```python
# Exploramos de nuevo los datos:
df_summary(st)
```
